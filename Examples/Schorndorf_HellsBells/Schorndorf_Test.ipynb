```python
def Load_U_Series_Ages(filename): 
    import pandas as pd
    df = pd.read_excel(filename)
    U_series_ages = df['U_ages'].values
    U_series_ages_err_low = df['U_Age_low'].values
    U_series_ages_err_high = df['U_Age_high'].values
    # Return ages and a tuple of their error bounds (low, high)
    return U_series_ages, (U_series_ages_err_low, U_series_ages_err_high)     
```

```python
ages, (ages_low, ages_high) = Load_U_Series_Ages('/Users/jackcarter/Initial Thorium /schorn_2018_IBIS_U_Series_Ages.xlsx')
```

```python
strat_inst = IBIS_strat.IBIS_Strat(ages, ages_low, ages_high, 
                                  schorn.df_reduced, sample_name = 'schorn_Testing', 
                                  iterations= 100000, burn_in=50000, 
                                  Start_from_pickles= True)
```

```python
thetas = strat_inst.Initial_Guesses_for_Model()
```
